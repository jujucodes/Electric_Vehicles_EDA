```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/EVDatabase.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# step 1: create integer values to use in EDA 
```

```python
df['PriceinGermany'] = df.PriceinGermany.str.replace('€', '')
df['PriceinGermany'] = df.PriceinGermany.str.replace(',', '')
df['PriceinGermany']
```

```python
df['PriceinGermany'] = pd.to_numeric(df['PriceinGermany'])
```

```python
df['PriceinUK'] = df.PriceinUK.str.replace('£', '')
df['PriceinUK'] = df.PriceinUK.str.replace(',', '')
df['PriceinUK']
```

```python
df['PriceinUK'] = pd.to_numeric(df['PriceinUK'])
```

```python
df['Acceleration'] = df.Acceleration.str.replace('sec', '')
df['Acceleration']
```

```python
df['Acceleration'] = pd.to_numeric(df['Acceleration'])
```

```python
df['TopSpeed'] = df.TopSpeed.str.replace('km/h', '')
df['TopSpeed']
```

```python
df['TopSpeed'] = pd.to_numeric(df['TopSpeed'])
```

```python
df['Range'] = df.Range.str.replace('km', '')
df['Range']
```

```python
df['Range'] = pd.to_numeric(df['Range'])
```

```python
df['Efficiency'] = df.Efficiency.str.replace('Wh/km', '')
df['Efficiency']
```

```python
df['Efficiency'] = pd.to_numeric(df['Efficiency'])
```

```python
df['FastChargeSpeed'] = df.FastChargeSpeed.str.replace('km/h', '')
df['FastChargeSpeed'] = df.FastChargeSpeed.str.replace('-', '')
df['FastChargeSpeed']
```

```python
df['FastChargeSpeed'] = pd.to_numeric(df['FastChargeSpeed'])
```

Hypothesis:

UK and German prices for Electric Vehicles follow very similar linear patterns. Both countries show a correlation between acceleration and price (with a faster acceleration associated with a higher market price) and range and price (larger ranges are equated to higher price). Tesla Roadster is an outlier with the highest price for UK and Germany and the farthest range and fastest acceleration, far outperforming other models on the market. 

If stakeholders are interested in pricing their Electric Vehicles higher when marketing to German and UK markets, a campaign focusing on the range and charging times of the vehicle and showcasing the ease of use for road trips and long commutes can help assuage consumer worries about driving longer distances and charging on the way.
